```python
import re


string = re.sub("\제","",nouns_list)
print(string)
```

```python
wc = WordCloud(width = 1000, height = 600, background_color="white", font_path=font_path)
plt.figure(figsize=(10, 10))
plt.imshow(wc.generate(string))
plt.axis("off")
plt.show() 
```

    - 정, 바, 때, 관, 정, 수, 그, 위, 거나, .. -> 의미없는 단어 집합

```python
string = re.sub("정","",string)
```

```python
string = re.sub("바","",string)
string = re.sub("때","",string)
string = re.sub("관","",string)
string = re.sub("그","",string)
string = re.sub("수","",string)
string = re.sub("위","",string)
string = re.sub("거나","",string)
```

```python
wc = WordCloud(width = 1000, height = 600, background_color="white", font_path=font_path)
plt.figure(figsize=(10, 10))
plt.imshow(wc.generate(string))
plt.axis("off")
plt.show()
```
