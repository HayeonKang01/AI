# 네이버 스마트스토어 품목, 가격 크롤링

```python
res = requests.get('https://smartstore.naver.com/yangpumso/category/6e6a8e6a3d934e7ebe04e1d0549f5846?cp=1')
res
```

```python
soup = BeautifulSoup(res.content, 'html.parser')
soup
```

```python
mydata = soup.find_all('strong', 'QNNliuiAk3')
name = []
for item in mydata:
    name.append(item.get_text())
```

```python
mydata = soup.find_all('span', 'nIAdxeTzhx')
all_price = []
for item in mydata:
    all_price.append(item.get_text())
```

```python
price = []
for i in range(0, len(all_price)):
    if i%2==0:
        price.append(all_price[i])
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'name':name, 
            'price':price})
```
